## Meetups type formating

```python
import os
import pandas as pd
```

```python
# reading every CSV with indexed sentences
# return a list object of files in the given folder
files_list = [f for f in os.listdir('extractedMeetupTypes') if not f.startswith('.')]
# parse to dataframe
df_files = pd.DataFrame(files_list, columns=['file_name'])
print(len(df_files))
# df_files = df_files.query("file_name=='10085.csv'")

df_files.info()
df_files.head()
```

```python
# Reading the biographies in chunks of 50
# for chunk in pd.read_csv('coreferenceBiographiesList.csv', chunksize=50):
# for chunk in pd.read_csv('list_wikiIdSample.csv', chunksize=50):
# for chunk in pd.read_csv('totalBiographiesBenchmark.csv', chunksize=50):
    # df_file_name = pd.DataFrame()
    # df_file_name['file_name'] = chunk['file_name']
    # # for each file
for file_name_item in df_files.itertuples():
    file_exists = os.path.isfile('extractedMeetupTypes/'+file_name_item.file_name)
    if file_exists:
        print(file_name_item.file_name.replace(".csv",""))
        # read the biography organised by sentences
        main_df = pd.read_csv('extractedMeetupTypes/'+file_name_item.file_name)
        if os.path.isfile('indexedSentences/'+file_name_item.file_name):
            results_df = pd.read_csv('indexedSentences/'+file_name_item.file_name)
            themes_df = main_df.loc[(main_df['Sentence'].isin(results_df.sentences.unique()))]
            values_df = themes_df[themes_df.columns[2:8]].copy()
            # values_df.info()
            themes_df['themeType'] = values_df.idxmax(axis=1).copy()
            themes_df['value'] = values_df.max(axis=1).copy()
            themes_df['sentenceIndex'] = results_df['sentenceIndex'].copy()
            themes_df['paragraphIndex'] = results_df['paragraphIndex'].copy()
            themes_df.drop(['Unnamed: 0'], inplace=True, axis=1)
            
            themes_df.to_csv('extractedMeetupTypes/'+file_name_item.file_name,index=False)
            # print(file_name_item.file_name,"Themes",len(themes_df))

        # store results
        # meetupsEval_df.to_csv('meetupsCandidatesAftCoref/'+file_name_item.file_name.replace(".txt",".csv"),index=False)
```
